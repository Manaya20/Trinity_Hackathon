```python
df_hcp_count = df.groupby('product name')['rendering provider doctor id'].nunique().reset_index()

total_hcp = df_hcp_count['rendering provider doctor id'].sum()
df_hcp_count['Percentage'] = (df_hcp_count['rendering provider doctor id'] / total_hcp) * 100


plt.pie(df_hcp_count['Percentage'], labels=df_hcp_count['product name'], autopct='%1.2f%%', labeldistance=1, radius=9)
plt.title('Percentage of HCPs Associated with Different Drugs')
plt.show()
```

```python
df['claim date'] = pd.to_datetime(df['claim date'], format='%Y-%m-%d')
```

```python
df['service bill amount'] = df['service bill amount'].replace('', np.nan)
```

```python
df['service bill amount_numeric'] = pd.to_numeric(df['service bill amount'], errors='coerce')
```

```python
non_numeric_rows = df[df['service bill amount_numeric'].isna()]
```

```python
print(non_numeric_rows[['service bill amount', 'service bill amount_numeric']])
```

```python

df['service bill amount'] = df['service bill amount'].replace('\xa0', '', regex=True)

```

```python

df['service bill amount'] = df['service bill amount'].replace('', np.nan)
df['service bill amount'] = df['service bill amount'].astype(float)
```

```python
df['service bill amount'] = df['service bill amount'].astype(str).str.replace('$', '').astype(float)
```

```python
df['service allowed amount'] = df['service allowed amount'].replace('\xa0', '', regex=True)
```

```python
df['service allowed amount'] = df['service allowed amount'].replace('', np.nan)
df['service allowed amount'] = df['service allowed amount'].astype(float)
```

```python
df['service allowed amount'] = df['service allowed amount'].astype(str).str.replace('$', '').astype(float)
```

```python
df['total_amount'] = df['service bill amount'] + df['service allowed amount']
```

```python
features= ['total_amount']
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])
```

```python
feature_2 = [' DoctorID']
```

```python
scaler = StandardScaler()
hcp_df_scaled = scaler.fit_transform(df_2[feature_2])
```
